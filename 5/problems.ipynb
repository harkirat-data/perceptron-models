```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input_1'] = [1,1,0,0]
or_data['input_2'] = [1,0,1,0]
or_data['output'] = [1,1,1,0]
```

```python
and_data['input_1'] = [1,1,0,0]
and_data['input_2'] = [1,0,1,0]
and_data['output'] = [1,0,0,0]
```

```python
xor_data['input_1'] = [1,1,0,0]
xor_data['input_2'] = [1,0,1,0]
xor_data['output'] = [0,1,1,0]
```

```python
from sklearn.linear_model import Perceptron
clf1 = Perceptron()
clf2 = Perceptron()
clf3 = Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
clf2.fit(or_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
clf3.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values)
```
